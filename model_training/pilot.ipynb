```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Class')
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=32)
```

```python
y_train
```

```python
from sklearn.pipeline import Pipeline
from category_encoders.target_encoder import TargetEncoder
from sklearn.decomposition import PCA
from xgboost import XGBClassifier

# Define the pipeline with TargetEncoder, PCA, and XGBClassifier
estimators = [
    ('pca', PCA()),  # You can adjust n_components as needed
    ('clf', XGBClassifier(random_state=32))  # Customize objective function with the objective parameter
]
pipe = Pipeline(steps=estimators)
pipe
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

search_space = {
    'clf__max_depth': Integer(2,8),
    'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
    'clf__subsample': Real(0.5, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode' : Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.0, 10.0),
    'clf__reg_lambda': Real(0.0, 10.0),
    'clf__gamma': Real(0.0, 10.0),
    'pca__n_components': Integer(2, 9)
}

opt = BayesSearchCV(pipe, search_space, cv=5, n_iter=12, scoring='roc_auc_ovr', random_state=32) 
# in reality, you may consider setting cv and n_iter to higher values
```

```python
opt.fit(X_train, y_train)
```

```python
opt.best_estimator_
```

```python
opt.best_score_
```

```python
opt.score(X_test, y_test)
```

```python
opt.predict(X_test)
```

```python
opt.predict_proba(X_test)
```
